```python
# ──────────────────────────────────────────────
# Q2. Load / Import Libraries (FIXED)
# ──────────────────────────────────────────────
import nltk
import spacy
from textblob import TextBlob
import string

# Force download all required NLTK resources
nltk.download('punkt')
nltk.download('punkt_tab')          # ← this was the missing one
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')  # ← sometimes needed too

print("✅ All NLTK resources downloaded")
```

```python
# ──────────────────────────────────────────────
# Q3. Tokenization using NLTK
# ──────────────────────────────────────────────
from nltk.tokenize import word_tokenize, sent_tokenize

sentence = "The quick brown fox jumps over the lazy dog. It was a sunny afternoon."

# Word Tokenization
word_tokens = word_tokenize(sentence)
print("📌 Original Sentence:")
print(sentence)

print("\n🔤 Word Tokens:")
print(word_tokens)
print(f"   Total word tokens: {len(word_tokens)}")

# Sentence Tokenization (bonus)
sent_tokens = sent_tokenize(sentence)
print("\n📄 Sentence Tokens:")
print(sent_tokens)
print(f"   Total sentences: {len(sent_tokens)}")
```

```python
# ──────────────────────────────────────────────
# Q4. POS Tagging using spaCy
# ──────────────────────────────────────────────

# Load English language model
nlp = spacy.load("en_core_web_sm")

sentence = "The quick brown fox jumps over the lazy dog."

# Process the sentence
doc = nlp(sentence)

print("📌 Sentence:", sentence)
print("\n🏷️  POS Tagging Results:")
print(f"{'Token':<15} {'POS':<10} {'Tag':<10} {'Description'}")
print("-" * 55)

for token in doc:
    print(f"{token.text:<15} {token.pos_:<10} {token.tag_:<10} {spacy.explain(token.tag_)}")
```

```python
# ──────────────────────────────────────────────
# Q5. Text Cleaning Function
# ──────────────────────────────────────────────

def clean_text(text: str) -> str:
    """
    Cleans input text by:
      1. Converting to lowercase
      2. Removing all punctuation

    Args:
        text (str): Raw input string
    Returns:
        str: Cleaned text
    """
    # Step 1: Lowercase
    text = text.lower()

    # Step 2: Remove punctuation using str.translate
    text = text.translate(str.maketrans('', '', string.punctuation))

    return text


# ── Test Cases ──
test_sentences = [
    "The Quick Brown Fox Jumps Over The Lazy Dog.",
    "Hello, World! How's it going?",
    "Data Science & AI/ML is AWESOME in 2025!",
    "Wait... Are you serious?! 100% confirmed."
]

print("🧹 Text Cleaning Demo\n")
print(f"{'Original':<45} → {'Cleaned'}")
print("-" * 75)

for text in test_sentences:
    cleaned = clean_text(text)
    print(f"{text:<45} → {cleaned}")
```
